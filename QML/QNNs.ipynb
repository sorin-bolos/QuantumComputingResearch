Machine learning tasks (supervized):

1. Classification: distinguish between different classes (cats vs dogs). Train os a set of labeled data so that when an unseen data record is presentet the program is able to say if it's a cat or a dog
2. Regression: find the function that best fits the data (find the correct price for a house). Train on a set of records with the known price so that when an unknown record is presented the program can return the correct price.

In ML there are two operations: 
1. Training - done once for creating a model
2. Inference - after the model is created use it to get correct results

There are two main methods in QML: Quantum Neural Networks and Quantum Kernels. 

We start with Quantum neural networks.

**NeuralNetwork**: The interface for neural networks. This is an abstract class all QNNs inherit from.

**EstimatorQNN**: A network based on the evaluation of quantum mechanical observables.

    Takes:
    * a circuit
    * a set of observables
    * input - this is the input data as parameters of the circuit
    * weights - parameters to optimize
    * an estimator instance (V1 or V2)

**SamplerQNN**: A network based on the samples resulting from measuring a quantum circuit.

    Takes:
    * a circuit
    * a function to post processthe sampler results
    * input - this is the input data as parameters of the circuit
    * weights - parameters to optimize
    * a ampler instance (V1 or V2)

## How to train

For Classification:

* NeuralNetworkClassifier \
  Takes: 
    * A QNN (EstimatorQNN or SamplerQNN)
    * An Optimizer function - function to perform optimization

For Regression:

* NeuralNetworkRegressor \
  Takes:
  * A QNN (EstimatorQNN or SamplerQNN)
  * A loss function (ex "squared_error")
  * An optimizer function

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
from qiskit.circuit.library import RealAmplitudes, ZZFeatureMap
from qiskit_machine_learning.optimizers import COBYLA, L_BFGS_B
from qiskit_machine_learning.utils import algorithm_globals

from qiskit_machine_learning.algorithms.classifiers import NeuralNetworkClassifier, VQC
from qiskit_machine_learning.algorithms.regressors import NeuralNetworkRegressor, VQR
from qiskit_machine_learning.neural_networks import SamplerQNN, EstimatorQNN
from qiskit_machine_learning.circuit.library import QNNCircuit

from qiskit.primitives import StatevectorEstimator
from qiskit.primitives import StatevectorSampler 

algorithm_globals.random_seed = 42
```

# Classification

```python
# Get the training data. We generate it
num_inputs = 2
num_samples = 20
X = 2 * algorithm_globals.random.random([num_samples, num_inputs]) - 1
y01 = 1 * (np.sum(X, axis=1) >= 0)  # in { 0,  1}
y = 2 * y01 - 1  # in {-1, +1}

for x, y_target in zip(X, y):
    print(f'Input: {x}.   Label: {y_target}')
```

```python
for x, y_target in zip(X, y):
    if y_target == 1:
        plt.plot(x[0], x[1], "bo")
    else:
        plt.plot(x[0], x[1], "go")
plt.plot([-1, 1], [1, -1], "--", color="black")
plt.show()
```

```python
# construct QNN with the QNNCircuit's default ZZFeatureMap feature map and RealAmplitudes ansatz.
qc = QNNCircuit(num_qubits=2)
qc.draw("mpl", style="clifford")
```

## Use Estimator QNN

```python
# Create the QNN
estimator = StatevectorEstimator()
estimator_qnn = EstimatorQNN(circuit=qc, estimator=estimator)
```

```python
# Train the QNN
estimator_classifier = NeuralNetworkClassifier(estimator_qnn, optimizer=COBYLA(maxiter=60))
estimator_classifier.fit(X, y)
```

```python
# Do predictions. Here we use the same data to see how the trained QNN performs
y_predict = estimator_classifier.predict(X)

for x, label, predicted in zip(X, y, y_predict):
    print(f'Input: {x}.      Label: {label},     Predicted Label: {predicted}')
```

```python
#Highlight misslabeled data

for x, y_target, y_p in zip(X, y, y_predict):
    if y_target == 1:
        plt.plot(x[0], x[1], "bo")
    else:
        plt.plot(x[0], x[1], "go")
    if y_target != y_p:
        plt.scatter(x[0], x[1], s=200, facecolors="none", edgecolors="r", linewidths=2)
plt.plot([-1, 1], [1, -1], "--", color="black")
plt.show()
```

## Use Sampler QNN

```python
# parity maps bitstrings to 0 or 1
def parity(x):
    return "{:b}".format(x).count("1") % 2


output_shape = 2 
```

```python
sampler = StatevectorSampler()
# construct QNN
sampler_qnn = SamplerQNN(
    circuit=qc,
    interpret=parity,
    output_shape=output_shape,
    sampler=sampler,
)
```

```python
# train
sampler_classifier = NeuralNetworkClassifier(neural_network=sampler_qnn, optimizer=COBYLA(maxiter=30))
sampler_classifier.fit(X, y01)
```

```python
# Do predictions. Here we use the same data to see how the trained QNN performs
y_predict = estimator_classifier.predict(X)

for x, label, predicted in zip(X, y, y_predict):
    print(f'Input: {x}.      Label: {label},     Predicted Label: {predicted}')
```

```python
#Highlight misslabeled data

for x, y_target, y_p in zip(X, y, y_predict):
    if y_target == 1:
        plt.plot(x[0], x[1], "bo")
    else:
        plt.plot(x[0], x[1], "go")
    if y_target != y_p:
        plt.scatter(x[0], x[1], s=200, facecolors="none", edgecolors="r", linewidths=2)
plt.plot([-1, 1], [1, -1], "--", color="black")
plt.show()
```

# Regression


```python
# Generate the data

num_samples = 20
eps = 0.2
lb, ub = -np.pi, np.pi
X_ = np.linspace(lb, ub, num=50).reshape(50, 1)
f = lambda x: np.sin(x)

X = (ub - lb) * algorithm_globals.random.random([num_samples, 1]) + lb
y = f(X[:, 0]) + eps * (2 * algorithm_globals.random.random(num_samples) - 1)

for x, y_target in zip(X, y):
    print(f'Input: {x}.   Label: {y_target}')
```

```python
plt.plot(X_, f(X_), "r--")
plt.plot(X, y, "bo")
plt.show()
```

```python
# Create circuit
# construct simple feature map
param_x = Parameter("x")
feature_map = QuantumCircuit(1, name="fm")
feature_map.ry(param_x, 0)

# construct simple ansatz
param_y = Parameter("y")
ansatz = QuantumCircuit(1, name="vf")
ansatz.ry(param_y, 0)

# construct a circuit
qc = QNNCircuit(feature_map=feature_map, ansatz=ansatz)
qc.draw('mpl')
```

```python
# construct QNN
regression_estimator_qnn = EstimatorQNN(circuit=qc, estimator=estimator)
```

```python
# train
regressor = NeuralNetworkRegressor(
    neural_network=regression_estimator_qnn,
    loss="squared_error",
    optimizer=L_BFGS_B(maxiter=5),
)
regressor.fit(X, y)
```

```python
# plot target function
plt.plot(X_, f(X_), "r--")

# plot data
plt.plot(X, y, "bo")

# plot fitted line
y_ = regressor.predict(X_)
plt.plot(X_, y_, "g-")
plt.show()
```
