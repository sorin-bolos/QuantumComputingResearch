```python
def get_sign_one_qubit(counts, h_counts, shots=1024):
    tolerance = 0.001
    
    zeros = counts.get('0', 0)
    ones = counts.get('1', 0)
    
    if zeros < shots*tolerance:
        return [0, 1]
    
    if ones < shots*tolerance:
        return [1, 0]
    
    difference = h_counts.get('0', 0) - h_counts.get('1', 0)    
    if difference > 0:
        return [1, 1]
    
    return [1, -1]
```

```python
def one_qubit_sign_test(theta):
    sv_simulator = BasicAer.get_backend('statevector_simulator')
    backend = BasicAer.get_backend('qasm_simulator')
    
    circuit = QuantumCircuit(1)
    circuit.ry(theta,0)
    statevector = execute(circuit, sv_simulator).result().get_statevector()
    
    measurement_circuit = QuantumCircuit(1)
    measurement_circuit.measure_all()
    circuit_with_measurement = circuit + measurement_circuit
    counts = execute(circuit_with_measurement, backend).result().get_counts(circuit_with_measurement)
    
    circuit.h(0)
    circuit_x_basis = circuit + measurement_circuit
    h_counts = execute(circuit_x_basis, backend).result().get_counts(circuit_x_basis)
    
    sign = get_sign_one_qubit(counts, h_counts)
    
    print(f'{statevector} ************** {sign}')
    
    return sign
```

```python
assert one_qubit_sign_test(0) == [1,0]
assert one_qubit_sign_test(pi) == [0,1]
assert one_qubit_sign_test(pi/2) == [1,1]
assert one_qubit_sign_test(-pi/2) == [1,-1]
assert one_qubit_sign_test(3*pi/2) == [1,-1]
assert one_qubit_sign_test(pi/3) == [1,1]
assert one_qubit_sign_test(2*pi/3) == [1,1]
assert one_qubit_sign_test(3*pi/3) == [0,1]
assert one_qubit_sign_test(4*pi/3) == [1,-1]
assert one_qubit_sign_test(5*pi/3) == [1,-1]
assert one_qubit_sign_test(6*pi/3) == [1,0]
assert one_qubit_sign_test(2.9*pi/3) == [1,1]
assert one_qubit_sign_test(-2.9*pi/3) == [1,-1]
```

```python
def two_independent_qubits_sign_test(theta0, theta1):
    sv_simulator = BasicAer.get_backend('statevector_simulator')
    backend = BasicAer.get_backend('qasm_simulator')
    
    circuit = QuantumCircuit(2)
    circuit.ry(theta0,0)
    circuit.ry(theta1,1)
    statevector = execute(circuit, sv_simulator).result().get_statevector()
    
    measurement_circuit = QuantumCircuit(2)
    measurement_circuit.measure_all()
    circuit_with_measurement = circuit + measurement_circuit
    result = execute(circuit_with_measurement, backend).result()
    counts = result.get_counts(circuit_with_measurement)
    
    circuit.h([0,1])
    circuit_x_basis = circuit + measurement_circuit
    result = execute(circuit_x_basis, backend).result()
    h_counts = result.get_counts(circuit_x_basis)
    
    def reduce_counts(counts, qubit_index):
        redued_counts = {'0':0, '1':0}
        for key in counts:
            redued_counts[key[-qubit_index-1]] += counts[key]
        return redued_counts
            
    counts_0 = reduce_counts(counts, 0)
    h_counts_0 = reduce_counts(h_counts, 0)        
    counts_1 = reduce_counts(counts, 1)
    h_counts_1 = reduce_counts(h_counts, 1)
    
    sign_0 = get_sign_one_qubit(counts_0, h_counts_0)
    sign_1 = get_sign_one_qubit(counts_1, h_counts_1)
    
    signs = [sign_0[0]*sign_1[0], sign_0[1]*sign_1[0], sign_0[0]*sign_1[1], sign_0[1]*sign_1[1]]

    print(f'{statevector} ************** {signs}')
    
    return signs
```

```python
assert two_independent_qubits_sign_test(0,0) == [1, 0, 0, 0]
assert two_independent_qubits_sign_test(pi,0) == [0, 1, 0, 0]
assert two_independent_qubits_sign_test(0,pi) == [0, 0, 1, 0]
assert two_independent_qubits_sign_test(pi,pi) == [0, 0, 0, 1]
assert two_independent_qubits_sign_test(pi/2,0) == [1, 1, 0, 0]
assert two_independent_qubits_sign_test(0,pi/2) == [1, 0, 1, 0]
assert two_independent_qubits_sign_test(pi/2,pi/2) == [1, 1, 1, 1]
assert two_independent_qubits_sign_test(-pi/2,-pi/2) == [1, -1, -1, 1]
assert two_independent_qubits_sign_test(pi/2,-pi/2) == [1, 1, -1, -1]
assert two_independent_qubits_sign_test(-pi/2,pi/2) == [1, -1, 1, -1]
assert two_independent_qubits_sign_test(-pi/3,0) == [1, -1, 0, 0]
assert two_independent_qubits_sign_test(0,-pi/3) == [1, 0, -1, 0]
assert two_independent_qubits_sign_test(pi/3,0) == [1, 1, 0, 0]
assert two_independent_qubits_sign_test(0,pi/3) == [1, 0, 1, 0]
assert two_independent_qubits_sign_test(2*pi/3,2*pi/3) == [1, 1, 1, 1]
assert two_independent_qubits_sign_test(-pi/3,-pi/3) == [1, -1, -1, 1]
assert two_independent_qubits_sign_test(pi/3,2*-pi/3) == [1, 1, -1, -1]
assert two_independent_qubits_sign_test(-pi/3,2*pi/3) == [1, -1, 1, -1]
assert two_independent_qubits_sign_test(-pi/3,0) == [1, -1, 0, 0]
assert two_independent_qubits_sign_test(0,2*-pi/3) == [1, 0, -1, 0]
```

```python
def two_entangled_qubits_sign_test(circuit):
    sv_simulator = BasicAer.get_backend('statevector_simulator')
    backend = BasicAer.get_backend('qasm_simulator')
    
    statevector = execute(circuit, sv_simulator).result().get_statevector()
    
    measurement_circuit = QuantumCircuit(2)
    measurement_circuit.measure_all()
    circuit_with_measurement = circuit + measurement_circuit
    result = execute(circuit_with_measurement, backend).result()
    counts = result.get_counts(circuit_with_measurement)
    
    circuit.h([0,1])
    circuit_x_basis = circuit + measurement_circuit
    result = execute(circuit_x_basis, backend).result()
    h_counts = result.get_counts(circuit_x_basis)
    
    def reduce_counts(counts, qubit_index):
        redued_counts = {'0':0, '1':0}
        for key in counts:
            redued_counts[key[-qubit_index-1]] += counts[key]
        return redued_counts
            
    counts_0 = reduce_counts(counts, 0)
    h_counts_0 = reduce_counts(h_counts, 0)        
    counts_1 = reduce_counts(counts, 1)
    h_counts_1 = reduce_counts(h_counts, 1)
    
    sign_0 = get_sign_one_qubit(counts_0, h_counts_0)
    sign_1 = get_sign_one_qubit(counts_1, h_counts_1)
    
    signs = [sign_0[0]*sign_1[0], sign_0[1]*sign_1[0], sign_0[0]*sign_1[1], sign_0[1]*sign_1[1]]

    print(f'{statevector} ************** {signs}')
    
    return signs
```
